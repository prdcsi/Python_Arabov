```python
def is_prime(n):
    if n < 2:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for i in range(3, int(n**0.5) + 1, 2):
        if n % i == 0:
            return False
    return True

print(" ,    (2^(2x+1) + 1)/3  1 <= x <= 3:")
print("-" * 60)

for x in range(1, 4):
    value = (2**(2*x + 1) + 1) // 3
    if is_prime(value):
        print(f"x = {x:2d} -> {value}")
```

```python
import math

epsilon = 1e-3
total_sum = 0.0
n = 1

while True:
    factorial_n = math.factorial(n)
    factorial_2n = math.factorial(2 ** n)
    an = factorial_n / factorial_2n
    total_sum += an
    if an < epsilon:
        break
    n += 1

print(f"    {epsilon}: {total_sum:.6f}")
print(f" : {n}")
```

```python
def find_special_numbers():
    results = []

    for a in range(1, 10):
        for b in range(0, 10):
            if b == a:
                continue
            for c in range(0, 10):
                if c == a or c == b:
                    continue
                for d in range(0, 10):
                    if d == a or d == b or d == c:
                        continue

                    ab = 10 * a + b
                    cd = 10 * c + d

                    digit_sum = a + b + c + d

                    if ab - cd == digit_sum:
                        number = 1000*a + 100*b + 10*c + d
                        results.append(number)
                        print(f"{a}{b}{c}{d}: {ab} - {cd} = {ab - cd},  = {digit_sum}")

    return results

numbers = find_special_numbers()
print(f"\n : {numbers}")
```

```python
import math

def divide_fractions(A, B, C, D):

    if C == 0:
        raise ValueError("  : C    0.")


    numerator = A * D
    denominator = B * C


    gcd = math.gcd(numerator, denominator)


    num_reduced = numerator // gcd
    den_reduced = denominator // gcd

    return num_reduced, den_reduced


A, B = 3, 4
C, D = 2, 5

result_num, result_den = divide_fractions(A, B, C, D)
print(f"({A}/{B}) : ({C}/{D}) = {result_num}/{result_den}")
```

```python
def factorial_sum(n):

    if n < 1 or n > 20:
        raise ValueError("n      1  20")

    total_sum = 0
    factorial = 1

    for i in range(1, n + 1):
        factorial *= i
        total_sum += factorial
    return total_sum


n = int(input(" n (1  n  20): "))
try:
    result = factorial_sum(n)
    print(f" 1! + 2! + ... + {n}! = {result}")
except ValueError as e:
    print(e)
```
